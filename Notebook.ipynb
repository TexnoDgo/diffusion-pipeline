```python
from diffusers import DiffusionPipeline
import torch
import transformers
```

```python
# Створення pipeline'у
pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipeline.to("cuda")

# Передача pipeline'y опису зображення
pipeline("An image of banana in Banksy style").images[0]
```
